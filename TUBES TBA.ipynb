```python
def tahap1(word):
    aray = []
    i = 0
    while(i<len(word)):
        temp = ''
        if(word[i] == 'p' or word[i] == 'q' or word[i] == 'r' or word[i] == 's'):
            temp = word[i]
            
        elif(word[i] == 'n'):
            try:
                if(word[i+1] == 'o'):
                    if(word[i+2] == 't'):
                        temp = 'not'
                        
                        i+=2

                    else:
                        aray.append('Error')
                        break
                else:
                    aray.append('Error')
                    break
            except:
                aray.append('Error')
                break
        elif(word[i] == 'a'):
            try:
                if(word[i+1] == 'n'):
                    if(word[i+2] == 'd'):
                        temp = 'and'
                        
                        i+=2
                    else:
                        aray.append('Error')
                        break
                else:
                    aray.append('Error')
                    break
            except:
                aray.append('Error')
                break
        elif(word[i] == 'o'):
            try:
                if(word[i+1] == 'r'):
                    temp = 'or'
                    
                    i+=1
                else:
                    aray.append('Error')
                    break
            except:
                aray.append('Error')
                break
        elif(word[i] == 'x'):
            try:
                if(word[i+1] == 'o'):
                    if(word[i+2] == 'r'):
                        temp = 'xor'
                        
                        i+=2
                    else:
                        aray.append('Error')
                        break
                else:
                    aray.append('Error')
                    break
            except:
                aray.append('Error')
                break
        elif(word[i] == 't'):
            try:
                if(word[i+1] == 'h'):
                    if(word[i+2] == 'e'):
                        if(word[i+3] == 'n'):
                            temp = 'then'
                            
                            i+=3
                        else:
                            aray.append('Error')
                            break
                    else:
                        aray.append('Error')
                        break
                else:
                    aray.append('Error')
                    break
            except:
                aray.append('Error')
                break
        elif(word[i] == 'i'):
            if(word[i+1] == 'f'):
                try:
                    if(word[i+2] == 'f'):
                        temp = 'iff'
                        
                        i+=2
                    elif(word[i+2] == ' '):
                        temp = 'if'
                        
                        i+=1
                    else:
                        aray.append('Error')
                        break
                except:
                    aray.append('Error')
                    break
            else:
                aray.append('Error')
                break
        elif(word[i] == ')'):
            temp = ')'
            
        elif(word[i] == '('):
            temp = '('
            
        
        try:
            if(word[i+1] == ' '):
                aray.append(temp)
                i+=2
            elif(word[i+1] == ')'):
                aray.append(temp)
                i+=1
            elif(word[i+1] == '('):
                aray.append(temp)
                i+=1
            elif(word[i] == '('):
                aray.append(temp)
                i+=1
            else:
                aray.append('Error')
                break
        except:
            if(i < len(word) and temp != ''):
                aray.append(temp)
                i+=1
            else:
                aray.append('Error')
                break
                
    token = {
        'p' : 1,
        'q' : 1,
        'r' : 1,
        's' : 1,
        'not' : 2,
        'and' : 3,
        'or' : 4,
        'xor' : 5,
        'if' : 6,
        'then' : 7,
        'iff' : 8,
        '(' : 9,
        ')' :10
    }
    
    arr = []
    
    for word in aray:
        if token.get(word) != None:
            arr.append(token.get(word))
        else:
            arr.append('Error')
            break
    
    return arr
```

```python
text_1 = "p and q or r"
text_2 = "if p and q then (not q s)"
text_3 = "p xor (q and not(p and q))"
text_4 = "(p and q ifg(r or s)"

arr1 = tahap1(text_1)
print(arr1)
arr2 = tahap1(text_2)
print(arr2)
arr3 = tahap1(text_3)
print(arr3)
arr4 = tahap1(text_4)
print(arr4)
```

```python
def tahap2(token):
    stack = []
    stack.append('#')
    x = 0
    while(x < len(token)):
        if(token[x] == 1):
            x+=1
            if(x == len(token) and  stack[len(stack)-1] == '#'):
                stack.clear()
                break;
            elif(token[x] != 3 and token[x] != 4 and token[x] != 5 and token[x] != 8 and token[x] != 7 and token[x] != 10):
                break
        elif(token[x] == 2):
            x += 1
            if(token[x] != 1 and token[x] != 9):
                break
        elif(token[x] == 3 or token[x] == 4 or token [x] == 5 or token[x] == 8):
            x+=1
            if(token[x] != 2 and token[x] != 1 and token[x] != 9):
                break
        elif(token[x] == 6) :
            x+=1
            stack.append("if");
        elif(token[x] == 7):
            closer = False
            while(stack[len(stack) - 1] != "#" and not closer) :
                if(stack[len(stack) - 1] == "if") :
                    closer = True
                stack.pop()    
            if (not closer) :
                break
            x += 1
        elif(token[x] == 9):
            x+=1
            stack.append("(")
        elif(token[x] == 10):
            x+=1
            closer = False
            while(stack[len(stack) - 1] != '#' and not closer):
                if(stack[len(stack) - 1] == "("):
                    closer = True
                stack.pop()
            if(not closer):
                break
            elif(x == len(token) and  stack[len(stack) - 1] == '#'):
                stack.clear()
                break

    if(len(stack) == 0):
        ay = 'valid'
    else:
        ay = 'Tidak Valid'
            
    return ay
```
